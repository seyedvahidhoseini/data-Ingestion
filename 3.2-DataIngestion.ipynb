```python
## Reading a PDf File
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('/content/drive/MyDrive/langchain_toturial/3.2-DataIngestion/attention.pdf')
docs=loader.load()
docs
```

```python
## Web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader=WebBaseLoader(web_paths=("https://www.khabaronline.ir/",),
                     bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                         class_=("container")
                     ))
                     )
```

```python
loader.load()
```
